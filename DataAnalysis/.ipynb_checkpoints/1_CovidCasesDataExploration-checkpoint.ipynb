## Reshaping Data 

```python
# Generating list of date columns and non date columns for easier filter
dtmask = []
cmask = []
for i in df.columns:
    if len(i.split("/")) == 3:
        dtmask.append(i)
    else:
        cmask.append(i)

# List of Location level Column namesS
locationmask = ["Country_Region", "Province_State", "Admin2"]
```

```python
# Prepare to generate Cases by states
flipmask = dtmask.copy()
flipmask.append("Province_State")
```

```python
# Cummulative cases over time in a Tableau friendly format
flipdf = df[flipmask]
flipdf = flipdf.groupby("Province_State").sum().transpose()

flipdf
```

```python
# Daily Cases by states
dailydf = flipdf.copy().diff()
dailydf.index = pd.to_datetime(dailydf.index)

dailydf
```

## Analytics

```python
# States with the most cases
# The top States will be selected to put into visualization

top10 = flipdf.loc['8/25/20',:].sort_values(ascending=False).head(10)
top10States = list(top10.index)

top10
```

```python
# Top 5 States trends

fig, axs = plt.subplots(1, 1, constrained_layout=True, figsize=(10,5))
fig.suptitle("Covid19 Cases Over Time")

axs.set_title("(Observation of cases spread around states)")

# Smoothened the data with a 2 weeks rolling window to make it look better
axs.plot(dailydf.rolling(14).mean().loc[:,top10States[0:4]])

monthyearFmt = mdates.DateFormatter('%Y-%m')
axs.xaxis.set_major_formatter(monthyearFmt)

axs.set_xlabel('Date')
axs.set_ylabel('Number of Cases')

axs.legend(top10States[0:4], loc='upper left')

plt.show()
```

```python
# Trending up states in the past two weeks

velocitydf = dailydf.iloc[-14:,:].diff()
velocitydf = velocitydf.transpose()

velocitydf['SumOfVelocityChanges'] = velocitydf.sum(axis=1)
```

```python
# Top 5 Virus Growing States
trendingUp = velocitydf.sort_values("SumOfVelocityChanges", ascending=False).head(5)

trendingUp['SumOfVelocityChanges']
```

```python
# Top 5 Virus Growing States

fig, axs = plt.subplots(1, 1, constrained_layout=True, figsize=(10,5))
fig.suptitle("Covid19 Cases Over Time")

axs.set_title("(Observation of currently surging states)")

# Smoothened the data with a 2 weeks rolling window to make it look better
axs.plot(dailydf.rolling(14).mean().loc[:,trendingUp.index])

monthyearFmt = mdates.DateFormatter('%Y-%m')
axs.xaxis.set_major_formatter(monthyearFmt)

axs.set_xlabel('Date')
axs.set_ylabel('Number of Cases')

axs.legend(trendingUp.index, loc='upper left')

plt.show()
```

## Porting out data to SQL

```python
# Creating columns for the table

AddColumn_Query = "ALTER TABLE CovidCase ADD COLUMN '{State}' REAL;"

queries = []
for i in dailydf.columns:
    queries.append(AddColumn_Query.format(State=i))
    
with open("CreateCovidCaseCol.sql", "w") as output:
    output.write("\n".join(queries))
```

```python
# Preparing the long column names to append into the SQL query

string = ""

for i in dailydf.columns:
    string = string+"'"+i+"'"+","

string = string[:-1]
```

```python
# Forming queries

Insert_Query = "INSERT INTO CovidCase (DATE) VALUES ('{date}');"
Insert_Query2 = "INSERT INTO CovidCase ({string}) VALUES({values});"

queries = []
for index, row in dailydf.iterrows():
    date = index.strftime("%Y-%m-%d")
    values = str(row.tolist())
    values = values[1:-1]
    queries.append(Insert_Query.format(date=date))
    queries.append(Insert_Query2.format(string=string, values=values))
    
with open(datetime.now().strftime("%Y%m%d")+"_CovidCaseInsert.sql", "w") as output:
            output.write("\n".join(queries))
```
